```python
#@title Importing the Data File
#At this point, open the dataset in google sheets or something and download as a pdf.
CMUF = pd.read_csv('/content/gdrive/MyDrive/Research/Typing Projects/Typing Public Dataset/CMUFree/CMUF.csv')
```

```python
#@title Generating Dataset with Words
# Create an empty DataFrame to store the transformed data
CMUW = pd.DataFrame(columns=['subject', 'screenIndex', 'word', 'key_presses'])

# Initialize an empty list to store key presses for each word
key_presses = []

# Iterate over each row in the data
for index, row in CMUF.iterrows():
    # If the current key press is a space, add the current word and its key presses to the transformed data
    if row['key2'] == 'space':
        word = ''.join(key_presses)
        CMUW = pd.concat([CMUW, pd.DataFrame({'subject': [row['subject']], 'screenIndex': [row['screenIndex']], 'word': [word], 'key_presses': [key_presses]})], ignore_index=True)
        key_presses = []
    else:
        # Add the current key press to the list of key presses for the current word
        key_presses.append(row['key2'])

# Add the last word and its key presses to the transformed data
word = ''.join(key_presses)
CMUW = pd.concat([CMUW, pd.DataFrame({'subject': [row['subject']], 'screenIndex': [row['screenIndex']], 'word': [word], 'key_presses': [key_presses]})], ignore_index=True)

# Display the transformed data
type(CMUW) #what kind of data is this?
#get summary about data frame currently active
CMUW.info()
#get subset of the dataset to see what it looks like
CMUW[:5]
```
